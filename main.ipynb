```python
import numpy as np
from gxgp import DagGP, Node, gp_common
import operator
from d3584 import f
```

```python
# Laod data
problem = np.load('problem_0.npz')
x_train, y_train = problem['x'], problem['y']
```

```python
# Model configuration
def safe_divide(a, b):
    if b == 0:
        return 0  # O qualche altro valore predefinito, a seconda del contesto
    return a / b
operators=[operator.add, operator.sub, operator.mul, safe_divide]
variables=[f'x{i}' for i in range(x_train.shape[0])]
constants=[1,2,3]
gp=DagGP(operators=operators, variables=variables,constants=constants)


population = [gp.create_individual(n_nodes=np.random.randint(5, 15)) for _ in range(100)]

```

```python
#Evolution
for generation in range(50):

    fitness = [gp.mse(ind, x_train.T, y_train) for ind in population]

    for ind, fit in zip(population, fitness):
        ind.fitness = fit

    for i, ind in enumerate(population):
        ind.index = i

    sorted_population = sorted(population, key=lambda ind: (ind.fitness, ind.index))
    # print("Popolazione ordinata:")
    # for ind in sorted_population:
    #     print(ind.fitness)

    # Selection and crossover
    offspring=[
        gp_common.xover_swap_subtree(sorted_population[i], sorted_population[i+1])
        for i in range (0, len(sorted_population)-1, 2)
    ]

    # Mutation
    for i in range(len(offspring)):
        if np.random.rand()<0.1:
            offspring[i]=gp.create_individual(n_nodes=5)


    fitness_offspring = [gp.mse(ind, x_train.T, y_train) for ind in offspring]

    for ind, fit in zip(offspring, fitness_offspring):
        ind.fitness = fit
    combined_population = sorted_population[:50] + offspring

    for i, ind in enumerate(combined_population):
        ind.index = i  

    population = sorted(combined_population, key=lambda ind: (ind.fitness, ind.index))


    
    print(f'Generation {generation}:best MSE={gp.mse(population[0], x_train.T, y_train)}')

# Best solution
best_individual=population[0]
print(f"Best formula found: {best_individual.long_name}")
```
